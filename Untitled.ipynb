# Импорт библиотек

```python
import pandas as pd
```

# Загрузка данных

```python
df_titanic = pd.read_csv('train.csv')
```

```python
df_titanic
```

# Предобработка данных - поиск и удаление NaN значений

### Поиск NaN значений:

```python
df_titanic.isnull().sum()
```

### Удаление NaN значений:

```python
df_titanic_without_NaN = df_titanic.dropna()
df_titanic_without_NaN
```

### Проверяем, что NaN значения удалены:

```python
df_titanic_without_NaN.isnull().sum()
```

# Фильтрация данных

### Добавление столбца для группировки по возрасту:

```python
dfg = df_titanic_without_NaN
```

```python
dfg['age_group'] = dfg.Age.apply(lambda x: '<=30' if x<=30 else '>30')
```

```python
dfg
```

### Группировка людей по полу и возрасту. Определение доли выживших в каждой группе:

```python
df_plot = dfg.groupby(['Sex','age_30']).Survived.mean()
df_plot
```

# Визуализация данных

```python
df_plot.plot.bar()
```

# Выводы:

### Больше всего выживших наблюдается среди женщин старше 30 лет. Меньше всего выжило мужчин старше 30 лет.
