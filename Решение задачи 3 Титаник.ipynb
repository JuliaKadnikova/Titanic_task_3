# 0. Импорт библиотек

```python
import pandas as pd
```

# 1. Загрузка данных

```python
df_titanic = pd.read_csv('train.csv')
```

```python
df_titanic
```

# 2. Предобработка данных - поиск и удаление NaN значений

### Поиск NaN значений:

```python
df_titanic.isnull().sum()
```

### Удаление NaN значений:

```python
df_titanic_without_NaN = df_titanic.dropna()
df_titanic_without_NaN
```

### Проверяем, что NaN значения удалены:

```python
df_titanic_without_NaN.isnull().sum()
```

# 3. Фильтрация данных

### Добавление столбца для группировки по возрасту:

```python
dfg = df_titanic_without_NaN
```

```python
dfAg = dfg.copy()
dfAg['age_group'] = dfg['Age'].apply(lambda x: '<=30' if x<=30 else '>30')
dfAg
```

### Группировка людей по полу и возрасту. Определение доли выживших в каждой группе:

```python
df_plot = dfAg.groupby(['Sex','age_group']).Survived.mean()
df_plot
```

# 4. Визуализация данных

```python
df_plot.plot.bar()
```

# 5. Выводы:

### Выживаемость выше среди женщин. Как видим на графике, больше всего выжило женщин старше 30 лет. Второе место по числу выживаемости заняли женщины 30 и младше лет. Чем это можно объяснить? Вероятнее всего, тем, что женщин и детей пытались спасти в первую очередь. Далее видим, что на третье место по числу выживших занимают мужчины 30 и младше лет, а вот мужчины старше 30 лет показывают самый низкий процент выживаемости. Почему? Вероятнее всего, дело в физической подготовке, силе молодого организма и том, что детей и женщин пытались спасти в первую очередь. Для более глубокого анализа можно построить и другие графики.
